# Custom trained object detection

* https://towardsdatascience.com/build-a-custom-trained-object-detection-model-with-5-lines-of-code-713ba7f6c0fb

```python
from detecto import utils, core, visualize
import numpy as np

image = utils.read_image("test_image.jpeg")
image = np.array(image)
```

```python
model = core.Model()
```

```python
labels, boxes, scores = model.predict_top(image)
```

```python
visualize.show_labeled_image(image,labels, boxes)
```
